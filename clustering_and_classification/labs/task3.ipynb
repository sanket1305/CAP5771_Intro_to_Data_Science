# Task 3: K-Means Clustering

## Overview

In this task, you will apply **K-Means clustering** to group similar documents together **without using labels.** This will help us explore whether documents naturally form meaningful clusters based on their text content.

## Why TF-IDF?

**K-Means** is a popular unsupervised learning algorithm that:

1. Groups documents based on their **similarity in feature space.**
2. Helps **discover patterns** in unlabeled text data.
3. Can be used to **compare with supervised classification results.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
def train_and_save_kmeans(k: int):
    # load tfdif train data
    X_train_tfidf = np.load('data/tfidf_train.npy')

    # generate kmeans model
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42).fit(X_train_tfidf)

    print("Training K-Means clustering with " + str(k) + " clusters")
    print("Clustering inertia:", kmeans.inertia_)

    return kmeans
```

## Elbow Method

The Elbow Method is a technique used in **KMeans clustering** to help determine the optimal number of clusters (k) for a given dataset. The goal is to find a balance between the number of clusters and the amount of variation explained by those clusters, so that you don't have too many clusters (which could lead to overfitting) or too few clusters (which could lead to underfitting).

```python
def elbow_method():
    inertia_values = []

    # for range of k_values extract inertia
    for k in range(5, 31):
        kmeans = train_and_save_kmeans(k)
        inertia_values.append(kmeans.inertia_)
    
    # Plot the elbow curve
    plt.plot(range(5, 31), inertia_values, marker='o')
    plt.title('Elbow Method for Optimal k')
    plt.xlabel('Number of clusters (k)')
    plt.ylabel('Inertia')
    plt.grid(True)
    # plt.show()
    plt.savefig('output/elbow_method.png')
```

```python
elbow_method()
```
